```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final_dataset.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
labels = ['Ddos', 'Benign'] ## 라벨 
fig = plt.figure(figsize=(8,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 배경색을 하얀색으로 설정
ax = fig.add_subplot() ## 프레임 생성
pie = ax.pie(df['Label'].value_counts(), ## 파이차트 출력
       startangle=90, ## 시작점을 90도(degree)로 지정
       autopct=lambda p : '{:.2f}%'.format(p), ## 퍼센티지 출력
       colors = ['purple', 'lightcoral'] ## 색상 지정
       )
 
plt.legend(pie[0],labels) ## 범례 표시
plt.show()
```

```python
df_sample = df.sample(frac=0.3, random_state=54)
```

```python
df_sample.shape
```

```python
df_sample.head()
```

```python
df_sample.isna().sum().head(40)
```

```python
df_sample.isna().sum().tail(44)
```

```python
df_sample.describe()
```

```python
df_sample1 = df_sample.dropna()
print(df_sample1.shape)
```

```python
df_sample1['Label'].value_counts()
```

```python
df_ddos = df_sample1[df_sample1['Label'] == 'ddos']
df_ddos.shape
```

```python
df_benign = df_sample1[df_sample1['Label'] == 'Benign']
df_benign.shape
```

```python
df_ddos_sample = df_ddos.sample(n=100000, random_state=54)
df_benign_sample = df_benign.sample(n=100000, random_state=54)
print(df_ddos_sample.shape)
print(df_benign_sample.shape)
```

```python
df_sampled = pd.concat([df_ddos_sample, df_benign_sample], axis=0)
df_sampled.shape
```

```python
df_sampled.head()
```

```python
df_final = df_sampled.sort_values(by='Timestamp')
```

```python
df_final.head(20)
```

```python
df_final["Label"].value_counts()
```

```python
df_final.to_csv('sampled_dataset.csv')
```

# ---------------------------------상관관계 파악---------------------------------------

```python
data = pd.read_csv('sampled_dataset.csv')
```

```python
onehot = pd.get_dummies(data['Label'])
onehot.head()
```

```python
data = data.join(onehot)
```

```python
data.head()
```

```python
data.columns
```

```python
cols = ['Flow_ID', 'Src_IP', 'Src_Port', 'Dst_IP', 'Dst_Port',
       'Protocol', 'Timestamp', 'Flow_Duration', 'Tot_Fwd_Pkts',
       'Tot_Bwd_Pkts', 'TotLen_Fwd_Pkts', 'TotLen_Bwd_Pkts', 'Fwd_Pkt_Len_Max',
       'Fwd_Pkt_Len_Min', 'Fwd_Pkt_Len_Mean', 'Fwd_Pkt_Len_Std',
       'Bwd_Pkt_Len_Max', 'Bwd_Pkt_Len_Min', 'Bwd_Pkt_Len_Mean',
       'Bwd_Pkt_Len_Std', 'Flow_Byts_per_s', 'Flow_Pkts_per_s',
       'Flow_IAT_Mean', 'Flow_IAT_Std', 'Flow_IAT_Max', 'Flow_IAT_Min',
       'Fwd_IAT_Tot', 'Fwd_IAT_Mean', 'Fwd_IAT_Std', 'Fwd_IAT_Max',
       'Fwd_IAT_Min', 'Bwd_IAT_Tot', 'Bwd_IAT_Mean', 'Bwd_IAT_Std',
       'Bwd_IAT_Max', 'Bwd_IAT_Min', 'Fwd_PSH_Flags', 'Bwd_PSH_Flags',
       'Fwd_URG_Flags', 'Bwd_URG_Flags', 'Fwd_Header_Len', 'Bwd_Header_Len',
       'Fwd_Pkts_per_s', 'Bwd_Pkts_per_s', 'Pkt_Len_Min', 'Pkt_Len_Max',
       'Pkt_Len_Mean', 'Pkt_Len_Std', 'Pkt_Len_Var', 'FIN_Flag_Cnt',
       'SYN_Flag_Cnt', 'RST_Flag_Cnt', 'PSH_Flag_Cnt', 'ACK_Flag_Cnt',
       'URG_Flag_Cnt', 'CWE_Flag_Count', 'ECE_Flag_Cnt', 'Down_per_Up_Ratio',
       'Pkt_Size_Avg', 'Fwd_Seg_Size_Avg', 'Bwd_Seg_Size_Avg',
       'Fwd_Byts_per_b_Avg', 'Fwd_Pkts_per_b_Avg', 'Fwd_Blk_Rate_Avg',
       'Bwd_Byts_per_b_Avg', 'Bwd_Pkts_per_b_Avg', 'Bwd_Blk_Rate_Avg',
       'Subflow_Fwd_Pkts', 'Subflow_Fwd_Byts', 'Subflow_Bwd_Pkts',
       'Subflow_Bwd_Byts', 'Init_Fwd_Win_Byts', 'Init_Bwd_Win_Byts',
       'Fwd_Act_Data_Pkts', 'Fwd_Seg_Size_Min', 'Active_Mean', 'Active_Std',
       'Active_Max', 'Active_Min', 'Idle_Mean', 'Idle_Std', 'Idle_Max',
       'Idle_Min', 'Benign', 'ddos']
```

```python
corr = data[cols].corr(method='pearson')
```

```python
plt.figure(figsize=(50,40))
sns.set(font_scale=1.5)
sns.heatmap(corr.values,
           cbar=True,
           annot=True,
           square=True,
           fmt='.2f',
           annot_kws={'size':10},
           yticklabels=cols,
           xticklabels=cols)
plt.tight_layout()
plt.show()
```
